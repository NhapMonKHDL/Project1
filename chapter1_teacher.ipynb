```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("heart_2022_no_nans.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
sub_df = df[["HeightInMeters", "WeightInKilograms", "BMI"]]
sub_df
```

```python
#Tinh trung binh va do lech chuan cua thuoc tinh HeightInMeters
mean_height = df["HeightInMeters"].mean()

std_height = np.std(df["HeightInMeters"])

#Tinh trung binh va do lech chuan cua thuoc tinh WeightInKilograms
mean_weight = df["WeightInKilograms"].mean()

std_weight = np.std(df["WeightInKilograms"])
```

```python
mean_weight, std_weight
```

```python
# Lay ra cac diem nam ngoai doan 3 sigma
sub_df[(sub_df["HeightInMeters"] < mean_height-std_height*3) | (sub_df["HeightInMeters"] > mean_height+std_height*3) 
       |((sub_df["WeightInKilograms"] < mean_weight-std_weight*3)) | (sub_df["WeightInKilograms"] > mean_weight+std_weight*3)]

```

```python
from scipy import stats
```

```python
##Chuan hoa du lieu bang zscore
#Cach 1: Su dung thu vien stats
sub_df["zscore_height_stats"] = stats.zscore(sub_df["HeightInMeters"])

#Cach 2: Tu tinh: (x-mean)/std
sub_df["zscore_height"] = [(x-mean_height)/std_height for x in df["HeightInMeters"]]

#Tinh zscore cua WeightInKilograms va BMI
sub_df["zscore_weight"] = stats.zscore(sub_df["WeightInKilograms"])

sub_df["zscore_BMI"] = stats.zscore(sub_df["BMI"])

sub_df
```

```python
#Lay cac diem outlier dua vao zscore
sub_df[(abs(sub_df["zscore_height"]) > 3) | (abs(sub_df["zscore_weight"]) > 3) | (abs(sub_df["zscore_BMI"]) > 3)]
```
